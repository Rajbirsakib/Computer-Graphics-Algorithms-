```python
import matplotlib.pyplot as plt

x1 = int(input("Enter value of x1: "))
y1 = int(input("Enter value of y1: "))
x2 = int(input("Enter value of x2: "))
y2 = int(input("Enter value of y2: "))

dx = x2-x1
dy = y2-y1
p0 = 2*dy-dx

if abs(dx)>abs(dy):
  steps = abs(dx)
else:
  steps = abs(dy)


xcor = []
ycor = []

i=0
while i<steps:
  i=i+1
  if p0>=0:
    x1=x1+1
    y1=y1+1
    p0=p0+2*dy-2*dx
  else:
    x1=x1+1
    y1=y1
    p0=p0+2*dy


  print("x1 : ",x1, "y1 :", y1)
  xcor.append(x1)
  ycor.append(y1)


plt.plot(xcor,ycor, color='orange', marker='o')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Bresenham Line Drawing Algorithm")
plt.show()
```
