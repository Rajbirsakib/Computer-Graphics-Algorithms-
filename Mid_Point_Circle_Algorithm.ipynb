```python
import matplotlib.pyplot as plt

x1 = int(input("Enter value of x1: "))
y1 = int(input("Enter value of y1: "))
r = int(input("Enter Radius r: "))

x0 = 0
y0 = r
Pk = 1-r

xcor=[]
ycor=[]

while x0<=y0:
  if Pk<0:
    x0=x0+1
    y0=y0
    Pk = Pk+2*x0 +1
  else:
    x0=x0+1
    y0=y0-1
    Pk = Pk+2*x0-2*y0+1

  #print("x:",x0, "y:",y0)
  xcor.append(x0)
  ycor.append(y0)
  xcor.append(y0)
  ycor.append(x0)
  xcor.append(x0)
  ycor.append(-y0)
  xcor.append(y0)
  ycor.append(-x0)
  xcor.append(-x0)
  ycor.append(-y0)
  xcor.append(-y0)
  ycor.append(-x0)
  xcor.append(-x0)
  ycor.append(y0)
  xcor.append(-y0)
  ycor.append(x0)

plt.scatter(xcor,ycor,color="orange")
#plt.gca().set_aspect('equal')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Mid Point Circle Algorithm")
plt.grid(True)
plt.show()
```
