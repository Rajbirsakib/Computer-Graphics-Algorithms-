```python
import matplotlib.pyplot as plt

x1 = int(input("Enter value of x1: "))
y1 = int(input("Enter value of y1: "))
x2 = int(input("Enter value of x2: "))
y2 = int(input("Enter value of y2: "))

dy = y2-y1
dx = x2-x1

m = dy/dx

if abs(dy)>abs(dx):
  steps=abs(dy)
else:
  steps=abs(dx)

xcor = []
ycor = []

i = 0

while i<steps:
  i+=1
  if m>1:
    x1=x1+(1/m)
    y1=y1+1
  elif m<1:
    x1=x1+1
    y1=y1+m
  else:
    x1=x1+1
    y1=y1+1

  print("x1: ",x1, "y1:",y1)
  xcor.append(x1)
  ycor.append(y1)


plt.plot(xcor,ycor,color='orange',marker='o')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("DDA Line Algorithm")
plt.show()
```
