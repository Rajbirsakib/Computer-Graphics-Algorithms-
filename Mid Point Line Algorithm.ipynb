```python
import matplotlib.pyplot as plt

x1 = int(input("Enter value of x1: "))
y1 = int(input("Enter value of y1: "))
x2 = int(input("Enter value of x2: "))
y2 = int(input("Enter value of y2: "))

dy = y2-y1
dx = x2-x1

Dk = (2*dy)-dx
dD = 2*(dy-dx)

if abs(dy)>abs(dx):
  steps=abs(dy)
else:
  steps=abs(dx)

xcor = []
ycor = []

i = 0
while i<steps:
  i+=1
  if Dk>=0:
    x1 = x1+1
    y1 = y1+1
    Dk = Dk+dD
  elif Dk<0:
    x1 = x1+1
    y1 = y1
    Dk = Dk+2*dy

  print("x1:",x1, "y1:",y1)
  xcor.append(x1)
  ycor.append(y1)

plt.plot(xcor,ycor,color='orange',marker='o')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Mid Point Line Algorithm")
plt.show()
```
